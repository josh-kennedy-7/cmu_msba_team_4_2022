```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, LogisticRegression
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
url = 'https://raw.githubusercontent.com/josh-kennedy-7/cmu_msba_team_4_2022/main/data/consolidated_tuna_data_clean.csv'
df = pd.read_csv(url,index_col=0,parse_dates=[0],thousands=',')
model_df = df.reset_index().copy()
model_df['year'] = pd.DatetimeIndex(model_df['year']).year

dataf = model_df.copy()
X_train, X_test, y_train, y_test = train_test_split(
    dataf.drop(labels=['skipjack_tuna_price','year','month'], axis=1),
    dataf['skipjack_tuna_price'],
    test_size=0.3,
    random_state=0)
X_train.shape, X_test.shape
scaler = StandardScaler()
scaler.fit(X_train.fillna(0))
sel_ = SelectFromModel(LogisticRegression(C=1, penalty='l2'))
sel_.fit(scaler.transform(X_train.fillna(0)), y_train)
selected_feat = X_train.columns[(sel_.get_support())]
```

```python
print("remaining features:")
for i in selected_feat:
  print(i)
```
